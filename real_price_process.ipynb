```python
# 引入依赖项
import csv
import os
import zipfile
import pandas as pd
from io import StringIO
```

```python
fromDir = "/home/coder/FX-1-Minute-Data/output"
toDir = '/home/coder/ADC2025/history_price'

productsList = [
    f for f in os.listdir(path=fromDir) if os.path.isdir(os.path.join(fromDir, f))
]

print(productsList)
```

```python
# 遍历每个产品文件夹，处理其中的zip文件
for product in productsList:
    product_dir = os.path.join(fromDir, product)
    zip_files = [f for f in os.listdir(product_dir) if f.endswith('.zip')]
    all_csv_data = []
    for zip_file in zip_files:
        zip_path = os.path.join(product_dir, zip_file)
        with zipfile.ZipFile(zip_path, 'r') as zf:
            for csv_name in zf.namelist():
                if csv_name.endswith('.csv'):
                    with zf.open(csv_name) as csvfile:
                        try:
                            # 自动检测分隔符，读取前6列，无header
                            sample = csvfile.read(2048).decode('utf-8')
                            csvfile.seek(0)
                            import csv as pycsv
                            sniffer = pycsv.Sniffer()
                            dialect = sniffer.sniff(sample)
                            df = pd.read_csv(csvfile, sep=dialect.delimiter, header=None, usecols=range(6), engine='python')
                            df.columns = ['time','open','high','low','close','volume']
                            all_csv_data.append(df)
                        except Exception as e:
                            print(f"读取{csv_name}出错: {e}")
    if all_csv_data:
        merged_df = pd.concat(all_csv_data, ignore_index=True)
        out_path = os.path.join(toDir, f"{product}.csv")
        merged_df.to_csv(out_path, index=False)
        print(f"已输出: {out_path}")
    else:
        print(f"{product} 没有找到csv数据")
```
